# OPENING WEBCAM

```python
def open_camera():
    # Tenta os índices e backends mais comuns no Windows.
    candidates = []
    for index in range(4):
        candidates.append((index, None))
        if hasattr(cv2, "CAP_DSHOW"):
            candidates.append((index, cv2.CAP_DSHOW))
        if hasattr(cv2, "CAP_MSMF"):
            candidates.append((index, cv2.CAP_MSMF))

    for index, backend in candidates:
        cap = cv2.VideoCapture(index) if backend is None else cv2.VideoCapture(index, backend)
        if cap.isOpened():
            print(f"Webcam aberta com sucesso no índice {index}")
            return cap
        cap.release()

    return None


def main():
    cap = open_camera()
    if cap is None:
        print("Erro: não foi possível acessar a webcam. Verifique permissão, driver e se outro app não está usando a câmera.")
        return

    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Erro: não foi possível capturar o frame")
                break

            frame = cv2.flip(frame, 1)

            cv2.imshow('Webcam', frame)
            key = cv2.waitKey(1) & 0xFF
            if key == 27:
                break
    finally:
        cap.release()
        cv2.destroyAllWindows()


main()
```
